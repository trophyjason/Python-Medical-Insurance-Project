Idea: find out who has a higher medical insurance charge, males or females

Goal: promote equal insurance costs for men and women

Data: sex and charges from medical insurance data

Analysis: using descriptive analysis to compare charges for females to males

```python
# Read csv file and created a list of dictionaries containing each row data with variable name as keys and row data as values
import csv
def read_file(filename):
    list_ = []
    with open(filename) as insurance_csv:
        insurance_file = csv.DictReader(insurance_csv)
        for row in insurance_file:
            list_.append(row)
    print("File: {}".format(list_))
    return list_
insurance_data = read_file('insurance.csv')
```

```python
# Created a function to create a list of dictionaries with needed keys
def list_of_needed_key_dicts(keys):
    list_ = []
    for data in insurance_data:
        dictionary = {}
        for key, value in data.items():
            for key_value in keys:
                if key == key_value:
                    dictionary[key_value] = value
        list_.append(dictionary)
    return list_
# Created list of dictionaries for only sex and charges variables
needed_keys = ["sex", "charges"]
sex_and_charges = list_of_needed_key_dicts(needed_keys)
print("List of dictionaries for sex and charge values for each person: {}".format(sex_and_charges))
```

```python
# Created a function that creates a dictionary with key values as keys and list of all data as value
def dict_labels_to_data(list_of_data, keys):
    dictionary = {}
    labels = []
    for values in list_of_data:
            if values[keys[0]] not in labels:
                labels.append(values[keys[0]])
    for label in labels:
        data = []
        for values in list_of_data:
            if values[keys[0]] == label:
                data.append(values[keys[1]])
        dictionary[label] = data
    return dictionary, labels
# Created a dictionary with sex values as keys and list of all charges as value
needed_keys_with_first_as_label = ["sex", "charges"]
sex_to_charges, labels_data = dict_labels_to_data(sex_and_charges, needed_keys_with_first_as_label)
print("Sex to list of charges dictionary: {}".format(sex_to_charges))
```

```python
# Created a function to find averages of each label for it's values and add to a dictionary
def averages_dict(list_of_data, labels):
    dictionary = {}
    for label in labels:
        total = 0
        length = 0
        for key, value in list_of_data.items():
            length = len(value)
            for charge in value:
                if key == label:
                    total += float(charge)
        average = total/length
        dictionary[label] = average
    return dictionary
# Created a dictionary for the averages of male and female charges
sex_avg_dict = averages_dict(sex_to_charges, labels_data)
print("Sex averages dictionary: {}".format(sex_avg_dict))
```

```python
# Finding which average is higher
def higher_value(dictionary):
    high = 0
    high_label = ""
    for key, value in dictionary.items():
        if value > high:
            high = value
            high_label = key
    return high_label, high
# Finding the higher average insurance charge between males and females
high_avg, high_value = higher_value(sex_avg_dict)
print("{} has the largest average insurance charge of {}".format(high_avg.title(), format(high_value, "0.2f")))
```

```python
def difference(dictionary):
    larger = ""
    smaller = ""
    values = []
    keys = []
    for key, value in dictionary.items():
        values.append(value)
        keys.append(key)
    value_1 = values[0]
    value_2 = values[1]
    key_1 = keys[0]
    key_2 = keys[1]
    if value_1 > value_2:
        larger = key_1
        smaller = key_2
        difference = value_1 - value_2
    elif value_2 > value_1:
        larger = key_2
        smaller = key_1
        difference = value_2 - value_1
    return difference, larger, smaller
diff, big, small = difference(sex_avg_dict)
print("{} has a higher average insurance charge compared to {} with a difference of {}".format(big.title(), small.title(), format(diff, "0.2f")))
```

Idea: find out who has a higher medical insurance charge, smokers or non-smokers

Goal: to stop people from smoking by significantly increasing insurance charges for smokers compared to non-smokers

Data: smoker and charges from medical insurance data

Analysis: using descriptive analysis to compare charges for smoker to non-smoker

```python
# Created a list of dictionaries with smoker and charges values
needed_keys_2 = ["smoker", "charges"]
smoker_and_charges = list_of_needed_key_dicts(needed_keys_2)
print("List of dictionaries for smoker and charge values for each person: {}".format(smoker_and_charges))
```

```python
# Created a dictionary with smoker values as keys and list of charges as values
needed_keys_with_first_as_label_2 = ["smoker", "charges"]
smoker_to_charges, labels_data_2 = dict_labels_to_data(smoker_and_charges, needed_keys_with_first_as_label_2)
print("Smoker to list of charges dictionary: {}".format(smoker_to_charges))
```

```python
# Created a dictionary with averages of smoker and non smoker charges
smoker_avg_dict = averages_dict(smoker_to_charges, labels_data_2)
print("Smoker average dictionary: {}".format(smoker_avg_dict))
```

```python
high_avg_2, high_value_2 = higher_value(smoker_avg_dict)
print("\"{}\" smoker has the largest average insurance charge of {}".format(high_avg_2.title(), format(high_value_2, "0.2f")))
```

```python
diff, big, small = difference(smoker_avg_dict)
print("\"{}\" smoker has a higher average insurance charge compared to \"{}\" smoker with a difference of {}".format(big.title(), small.title(), format(diff, "0.2f")))
```
